```python
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
%matplotlib widget
import numpy as np
```

```python
file_path = '/Users/lukas/Dokumente/Uni/Sentinel/Sentinel2_Germany_9728.tif'
```

```python
#file_path = '/path/to/your/tiff_file.tif'

# Open the TIFF file
with rasterio.open(file_path) as src:
    # Read the raster data
    data = src.read()  # Reads the first band
    data /= 10000 #conversion from DN
    data *= 2

    # Show the raster data
    #plt.show(data)
```

```python
red_img = data[3]
blue_img = data[1]
green_img = data[2]

rgb_img = np.stack([red_img, green_img, blue_img], axis=-1)
```

```python
plt.close('all')
plt.figure(figsize=(8,8))
plt.imshow(rgb_img)
plt.show()
```
